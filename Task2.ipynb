```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
import numpy as np
x = np.arange(0,100)
y = x*2
z = x**2
```

```python
#  Follow along with these steps: **

# ** Create a figure object called fig using plt.figure() **
# ** Use add_axes to add an axis to the figure canvas at [0,0,1,1]. Call this new axis ax. **
# ** Plot (x,y) on that axes and set the labels and titles to match the plot below:**

fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
x_end = 100
y_end = 200
m = y_end / x_end
xi = np.linspace(0, x_end, 100)
yi = m * xi
ax.plot(xi, yi)
ax.set_xlabel('X')
ax.set_ylabel('Y ')
ax.set_title('title')

ax.set_xlim([0, x_end])
ax.set_ylim([0, y_end])
ax.set_xticks(np.arange(0, x_end+1, 20))
ax.set_yticks(np.arange(0, y_end+1, 50))

plt.show()
```

```python

fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
x2 = np.arange(0,1.1,0.2)
y2 = np.arange(0,1.1,0.2)

ax2.set_xlabel('x')
ax2.set_ylabel('y')

ax2.set_xlim([0, 1]) #الحدود
ax2.set_ylim([0, 1])

ax2.set_xticks(x2)
ax2.set_yticks(y2)

# Display the plot
fig.show()
```

```python
fig = plt.figure()
ax1 = fig.add_axes([0, 0, 1, 1])
ax2 = fig.add_axes([0.2, 0.5, 0.2, 0.2])
x_end = 100
y_end = 200
m = y_end / x_end
x = np.linspace(0, x_end, 100)
y = m * x
ax1.plot(x, y,'r-')
ax2.plot(x, y,'r-')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')
ax1.set_xlabel('X')
ax1.set_ylabel('Y')
ax1.set_xlim([0, 100])
ax1.set_ylim([0, 200])
ax2.set_xlim([0, 100])
ax2.set_ylim([0, 200])
ax1.set_xticks(np.arange(0,101,20))
ax1.set_yticks(np.arange(0,201,50))
ax2.set_xticks(np.arange(0,101,20))
ax2.set_yticks(np.arange(0,201,50))
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(8, 6))

ax.plot(x, z)
ax2 = ax.inset_axes([0.2, 0.5, 0.4, 0.4])
ax2.plot(x, y, color='blue')
ax2.set_xlim(20, 22)  # تحديد حدود x
ax2.set_ylim(30, 50)  # تحديد حدود y
ax2.set_title('zoom')

ax.set_xlabel('X')
ax.set_ylabel('Z')
ax2.set_xlabel('X')
ax2.set_ylabel('Y')

plt.show()
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
#_________________________________________________________
fig, axes = plt.subplots(nrows=1, ncols=2)
axes[0].set_yticks(np.arange(0,201,50))
axes[0].set_xticks(np.arange(0,101,20))
axes[1].set_xticks(np.arange(0,101,20))
axes[0].plot(x, y,'b--',linewidth=2)
axes[1].plot(x, z, 'r-', linewidth=2)
plt.show()
```

```python
x = np.arange(0,100)
y = x*2
z = x**2
#_________________________________________________________
fig, axes = plt.subplots(nrows=1, ncols=2,figsize=(12,2))
axes[0].set_yticks(np.arange(0,201,50))
axes[0].set_xticks(np.arange(0,101,20))
axes[1].set_xticks(np.arange(0,101,20))
axes[0].plot(x, y,'b-',linewidth=4)
axes[1].plot(x, z, 'r--', linewidth=2)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
sns.set_style('whitegrid')
```

```python
titanic = sns.load_dataset('titanic')
```

```python
titanic.head()
```

```python
sns.jointplot(x='fare',y='age',data=titanic)
plt.xlim(-100,600,100)
```

```python
sns.histplot(titanic['fare'], bins=30, color='deeppink', label='Fare')
sns.histplot(titanic['age'], bins=30, color='deeppink',  label='')
plt.xlim(0, 600)
plt.show()
```

```python
plt.figure()

# رسم الصناديق (boxplot) للعمر مقابل الفئة
sns.boxplot(data=titanic, x='class', y='age', palette='Set2')

# تعيين التسميات والعنوان
plt.xlabel('Class')
plt.ylabel('age')
```

```python
sns.swarmplot(x='class',y='age',data=titanic,palette='rainbow')
```

```python
sns.countplot(data= titanic, x='sex', palette=['blue', 'green'])
plt.show()
```

```python
sns.heatmap(titanic.corr())
plt.show()
```
