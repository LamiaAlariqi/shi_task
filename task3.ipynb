```python
sns.FacetGrid(data, hue = "Survived").map(sns.distplot, "Age").add_legend()
plt.show()

# Among all the age groups, the children and people between 0 and 50  of age are the highest.
# There is a high overlap between the class labels. This implies that the survival 
#  of the people cannot be discerned from the person’s age.
```

```python
#رسم بياني لتوزيع الناجيين وغير الناجيين 
survived=data['Survived'].value_counts()
survived.index = ['Non-Survivors', 'Survivors']
plt.bar(survived.index,survived.values)
plt.xlabel('Survived')
plt.ylabel('Count')
plt.title('Survival Distrubtion')
plt.show()
# this diagram shows the amount of survival and non survival people 
#we can notice that the number of survival is half the number of non survival  
```

```python
#توزيع حسب الاعمار 
# رسمة توزيع الأعمار في قاعدة بيانات  توضح كيفية توزيع
# الأعمار للأشخاص الموجودين في القاعدة. يتم تمثيل المحور الأفقي 
# بفئات الأعمار المختلفة، بينما يتم تمثيل المحور العمودي  بعدد الأشخاص في كل فئة عمرية.
#الرسمة هنا توضح زيادة في عدد الاطفال 

plt.hist(data['Age'],bins=20)
plt.xlabel('Age')
plt.ylabel('Count')
plt.title('Age Distrubtion')
plt.show()
```

```python
#توزيع الركاب حسب الجنس 
#نلاحظ من الرسمة ان عدد الذكور المسافرين  يفوق عدد الاناث 
sex_count=data['Sex'].value_counts()
plt.pie(sex_count,labels=sex_count.index,autopct='%1.1f%%')
plt.title('passanger Gender Distribution')
plt.show()
```
